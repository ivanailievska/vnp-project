```python
import yfinance as yf
import pandas as pd
from tsfresh import extract_features
from tsfresh.utilities.dataframe_functions import impute
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
symbol = "^GSPC"
start_date = "2019-08-17"
end_date = "2023-08-17"

stock_data = yf.download(symbol, start=start_date, end=end_date)
```

```python
stock_data
```

```python
stock_data = stock_data[['Close']].reset_index()
stock_data.columns = ['Date', 'Close']
```

```python
stock_data
```

```python
extracted_features = extract_features(stock_data, column_id="Date", impute_function=impute)
```

```python
extracted_features
```

```python
plt.figure(figsize=(10, 6))
plt.plot(stock_data['Date'], stock_data['Close'], label='S&P 500')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('S&P 500 Stock Market Data')
plt.legend()
plt.tight_layout()

plt.show()
```

```python
fig = px.line(stock_data, x='Date', y='Close', title='S&P 500 Stock Market Data')
fig.update_xaxes(title_text='Date')
fig.update_yaxes(title_text='Close Price')
fig.show()
```
